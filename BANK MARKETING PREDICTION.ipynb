# Step 1: Loading the Data

```python
import pandas as pd

# Loading the training and test datasets
train_data = pd.read_csv("termdeposit_train.csv")
test_data = pd.read_csv("termdeposit_test.csv")
```

# STEP 4: Data Preprocessing ,Training ,Testing and Prediciton

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier


# Identifying categorical columns excluding 'subscribed'
categorical_cols = train_data.select_dtypes(include=['object']).drop(['subscribed'], axis=1).columns

# Instantiating LabelEncoder for features
label_encoder_features = LabelEncoder()

# Combining unique values from both training and test sets for features
combined_unique_values_features = set(train_data[categorical_cols].stack().unique())

# Fitting LabelEncoder on the combined unique values for features
label_encoder_features.fit(list(combined_unique_values_features))

# Applyig LabelEncoder to each categorical column in both the training set
for col in categorical_cols:
    train_data[col] = label_encoder_features.transform(train_data[col])

# Mapping 'yes' and 'no' to 1 and 0 for the target variable
target_mapping = {'yes': 1, 'no': 0}
train_data['subscribed'] = train_data['subscribed'].map(target_mapping)

# Splitting the dataset into features and target variable for training
X_train = train_data.drop(['ID', 'subscribed'], axis=1)
y_train = train_data['subscribed']


# Applying LabelEncoder to the same categorical columns in the test set
for col in categorical_cols:
    test_data[col] = label_encoder_features.transform(test_data[col])

# Dropping the 'ID' column from the test set
test_features = test_data.drop('ID', axis=1)

# Training the XGBoost model
model = XGBClassifier()
model.fit(X_train, y_train)

# Making predictions on the test set
predictions = model.predict(test_features)

# Mapping predictions back to 'yes' and 'no'
predicted_labels = ['yes' if pred == 1 else 'no' for pred in predictions]

# Adding predictions to the test dataset
test_data['subscribed'] = predicted_labels

# Printing the predictions
print("Predictions:", predicted_labels)

# Saving the predictions to a CSV file
test_data[['ID', 'subscribed']].to_csv('term_deposit_predictions.csv', index=False)
```

```python
# Plotting the predictions with counts
plt.figure(figsize=(8, 6))
subscription_counts = test_data['subscribed'].value_counts()
subscription_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Term Deposit Subscription Predictions')
plt.xlabel('Subscription Status')
plt.ylabel('Count')

# Annotating the bars with counts
for i, count in enumerate(subscription_counts):
    plt.text(i, count + 50, str(count), ha='center', va='bottom')

plt.xticks(rotation=0)
plt.show()
```

# Step 5: HYPERPARAMETERS TUNNING ON RANDOM FOREST MODEL

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt


# Identifying categorical columns
categorical_cols = train_data.select_dtypes(include=['object']).columns

# Removing 'subscribed' from categorical columns as it's the target variable
categorical_cols = categorical_cols.drop(['subscribed'], errors='ignore')

# Instantiating LabelEncoder for features
label_encoder_features = LabelEncoder()

# Combining unique values from both training and test sets for features
combined_unique_values_features = set(train_data[categorical_cols].stack().unique())

# Fitting LabelEncoder on the combined unique values for features
label_encoder_features.fit(list(combined_unique_values_features))

# Applying LabelEncoder to each categorical column in both the training set
for col in categorical_cols:
    train_data[col] = label_encoder_features.transform(train_data[col])

# Mapping 'yes' and 'no' to 1 and 0 for the target variable
target_mapping = {'yes': 1, 'no': 0}
train_data['subscribed'] = train_data['subscribed'].map(target_mapping)

# Splitting the dataset into features and target variable for training
X_train = train_data.drop(['ID', 'subscribed'], axis=1)
y_train = train_data['subscribed']

# Defining the parameter grid for Grid Search
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Instantiating the Random Forest classifier
rf_model = RandomForestClassifier()

# Instantiating GridSearchCV with the Random Forest model and parameter grid
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=3, scoring='accuracy', n_jobs=-1)

# Performing Grid Search to find the best hyperparameters
grid_search.fit(X_train, y_train)

# Printing the best hyperparameters found
print("Best Hyperparameters:", grid_search.best_params_)

# Training the Random Forest model with the best hyperparameters
best_rf_model = grid_search.best_estimator_

# Making predictions on the test set
rf_predictions = best_rf_model.predict(X_train)

# Mapping predictions back to 'yes' and 'no'
rf_predicted_labels = ['yes' if pred == 1 else 'no' for pred in rf_predictions]

# Plotting the Random Forest predictions with counts
plt.figure(figsize=(8, 6))
rf_subscription_counts = pd.Series(rf_predicted_labels).value_counts()
rf_subscription_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Random Forest Term Deposit Subscription Predictions (Tuned)')
plt.xlabel('Subscription Status')
plt.ylabel('Count')

# Annotating the bars with counts
for i, count in enumerate(rf_subscription_counts):
    plt.text(i, count + 50, str(count), ha='center', va='bottom')

plt.xticks(rotation=0)
plt.show()
```

```python
from sklearn.metrics import accuracy_score

# Calculating accuracy on the training set
accuracy = accuracy_score(y_train, rf_predictions)

# Printing the accuracy percentage
print(f"ACCURACY OF BEST MODEL RANDOM FOREST IS : {accuracy * 100:.2f}%")
```

**Observation:**

**The Random Forest model, tuned with the best hyperparameters obtained through Grid Search, demonstrates a remarkable accuracy of 96.43%. This high accuracy percentage indicates that the model is effective in correctly predicting whether a client will subscribe to a term deposit.**
