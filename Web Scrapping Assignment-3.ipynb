```python
# 5.Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

# Define the function to scrape coordinates
def scrape_coordinates(city_name):
    try:
        # Start a new Chrome browser session
        driver = webdriver.Chrome()
        
        # Open Google Maps
        driver.get("https://www.google.com/maps")
        
        # Find the search box and enter the city name
        search_box = driver.find_element(By.XPATH, "//input[@id='searchboxinput']")
        search_box.send_keys(city_name)
        search_box.send_keys(Keys.RETURN)
        
        # Wait for the search results to load (you can adjust the sleep time if needed)
        time.sleep(5)
        
        # Get the current URL (it may contain the coordinates)
        current_url = driver.current_url
        
        # Parse the URL to extract the latitude and longitude
        latitude = None
        longitude = None
        
        # Extract coordinates from the URL (you may need to adjust the XPath)
        if "/@" in current_url:
            coordinates = current_url.split("/@")[1].split(",")
            latitude = coordinates[0]
            longitude = coordinates[1]
        
        print(f"City: {city_name}")
        if latitude and longitude:
            print(f"Latitude: {latitude}")
            print(f"Longitude: {longitude}")
        else:
            print("Coordinates not found.")
        
    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        driver.quit()

# List of cities to search
cities = [
    "New York",
    "Los Angeles",
    "London",
    "Paris",
    "Tokyo",
    "Sydney",
    "Toronto",
    "Rome",
    "Berlin",
    "Beijing"
]

# Loop through the cities and call the function to scrape coordinates
for city in cities:
    scrape_coordinates(city)
    print("=" * 20)
```

```python
#9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location.
#You have to scrape hostel name, distance from city centre, ratings, total reviews, 
#overall reviews, privates from price, dorms from price, facilities and property description.


import selenium
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
from selenium import webdriver
import time #use to stop search engine for few seconds
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as E

# Set up Selenium webdriver
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 10)
```

```python
driver.get("https://www.hostelworld.com/s?q=London, England&country=England&city=London&type=city&id=3&from=2022-05-25&to=2022-05-28&guests=2&page=1")
```

```python
#Extracting url
product_urls = []
url=driver.find_elements(By.XPATH,'//a[@class="property-card-container"]')       #scraping urls
for i in url:
    product_urls.append(i.get_attribute("href")) 
```

```python
len(product_urls)
```

```python
title = []
distance = []
rating = []
total_review = []
prices = []
private_price = []
dorms_price = []
facilities = []
prop_details = []
```

```python
for url in product_urls:
    driver.get(url)
    time.sleep(2)  # Added a sleep time

    try:
        title_tags = driver.find_elements(By.XPATH, '//div[@class="title-2"]')
        for i in title_tags:
            title.append(i.text)
    except NoSuchElementException:
        title.append('-')

    try:
        distance_tags = driver.find_elements(By.XPATH, '//div[@class="property-distance"]')
        for j in distance_tags:
            distance.append(j.text.split('-')[1])
    except NoSuchElementException:
        distance.append('-')

    try:
        rating_tags = driver.find_elements(By.XPATH, '//div[@class="score orange big"]')
        for k in rating_tags:
            rating.append(k.text)
    except NoSuchElementException:
        rating.append('-')

    try:
        total_review_tags = driver.find_elements(By.XPATH, '//div[@class="rating rating-summary-container big clickable"]')
        for l in total_review_tags:
            total_review.append(l.text)
    except NoSuchElementException:
        total_review.append('-')

    try:
        P_prices_tags = driver.find_elements(By.XPATH, '//div[1][@class="property-accommodation-price"]')
        for n in P_prices_tags:
            private_price.append(n.text)
    except NoSuchElementException:
        private_price.append('-')

    try:
        D_prices_tags = driver.find_elements(By.XPATH, '//div[2][@class="property-accommodation-price"]')
        for o in D_prices_tags:
            dorms_price.append(o.text)
    except NoSuchElementException:
        dorms_price.append('-')

    try:
        faci_tags = driver.find_elements(By.XPATH, '//ul[@class="facility-sections"]')
        for p in faci_tags:
            facilities.append(p.text)
    except NoSuchElementException:
        facilities.append('-')

    try:
        prop_des = driver.find_elements(By.XPATH,'//div[@class="content collapse-content"]')
        for q in prop_des:
            prop_details.append(q.text)
    except NoSuchElementException:
        prop_details.append('-')
```

```python
# Print the results
print("Titles:", title)
print("Distances:", distance)
print("Ratings:", rating)
print("Total Reviews:", total_review)
print("Private Prices:", private_price)
print("Dorms Prices:", dorms_price)
print("Facilities:", facilities)
print("Property Details:", prop_details)
```

```python
#8. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.


import selenium
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
from selenium import webdriver
import time #use to stop search engine for few seconds
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as E

# Set up Selenium webdriver
driver = webdriver.Chrome()
```

```python
driver.get("https://www.youtube.com/watch?v=TFr6G5zveS8")
```

```python
# To scroll down
for _ in range(1000):
    driver.execute_script('window.scrollBy(0,5000)')
    
comment = []
vote = []
time = []

# Scrapping Comment
try:
    comment_tag = driver.find_elements(By.XPATH,'//div[@class="style-scope ytd-comment-renderer"]')
    for i in comment_tag:
        comment.append(i.text)
except NoSuchElementException:
        comment.append("-")    

# Scrapping upvote
try:
    vote_tag = driver.find_elements(By.XPATH,'//span[@class="style-scope ytd-comment-action-buttons-renderer"]')
    for j in vote_tag:
        vote.append(j.text)
except NoSuchElementException:
        vote.append("-")

#Scrapping time
try:   
    time_tag = driver.find_elements(By.XPATH,'//a[@class="yt-simple-endpoint style-scope yt-formatted-string"]')
    for k in time_tag:
        time.append(k.text)
except NoSuchElementException:
        time.append("-")
```
